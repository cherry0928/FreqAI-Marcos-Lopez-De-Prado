```python
from freqtrade.configuration import Configuration
from freqtrade.data.btanalysis import load_backtest_data, load_backtest_stats

Path('user_data', 'strategies', 'VulcanPrimary.json')
config = Configuration.from_files([Path('user_data', 'strategies', 'VulcanExplore.json')])

# Automatically selects the latest backtest results if no file provided
backtest_dir = Path(config['user_data_dir'], 'backtest_results', 'backtest_vulcanprimary-2022-05-01_11-51-51.json')
```

```python
trades = load_backtest_data(backtest_dir)
```

```python
trades
```

```python
trades.sort_values(['pair', 'open_date'], inplace=False)
```

```python
# see biggest loosers
trades.sort_values('profit_ratio', inplace=True)
trades.head(30)
```

```python
worst_trades = trades[trades['profit_ratio'] < -0.1]
worst_trades
```

```python
sns.histplot(worst_trades['profit_ratio'], stat='percent')
```

```python
worst_trades.to_csv('vulcan_worst_trades.csv')
```
