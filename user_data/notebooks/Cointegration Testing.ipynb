```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pathlib import Path
from freqtrade.configuration import Configuration
from freqtrade.data.history import load_pair_history
from freqtrade.resolvers import StrategyResolver
from freqtrade.data.dataprovider import DataProvider
```

```python
# Initialize empty configuration object
config = Configuration.from_files([Path('user_data', 'strategies', 'CointegrationExplore.json')])

# Location of the data
data_location = Path(config['user_data_dir'], 'data', 'binance')
```

```python
temp = []

for pair in config['exchange']['pair_whitelist']:    
    df = load_pair_history(datadir=data_location,
                           timeframe=config["timeframe"],
                           pair=pair,
                           data_format = "json")
    
    df.rename(columns={'open': '%s_open' % pair}, inplace=True)
    df.drop(columns=['high', 'low', 'close', 'volume'], inplace=True)
    temp.append(df)

pairs_df = pd.concat(temp)
```

```python
btc = df['open']
eth = df['open_eth_5m']
```

```python
plt.figure(figsize=(16, 4))

btc_norm = 100 * btc / btc[0]
eth_norm = 100 * eth / eth[0]

plt.plot(btc_norm, label='BTC', linewidth=1)
plt.plot(eth_norm, label='ETH', linewidth=1)

plt.legend()
plt.show()
```

```python
coint = CointAnalysis()

X = np.stack([btc, eth], axis=1)

coint.test(X).pvalue_
```

```python
coint.fit(X)
print(f'coef: {coint.coef_}')
print(f'mean: {coint.mean_}')
print(f'std: {coint.std_}')
```

```python
plt.figure(figsize=(16, 4))

btc_adj = (-coint.coef_[0]) * btc + coint.mean_

plt.plot(btc_adj, label=f'{-coint.coef_[0]:.2f} * BTC + {coint.mean_:.2f}', linewidth=1)
plt.plot(eth, label='ETH', linewidth=1)

plt.legend()
plt.show()
```

```python
spread = coint.transform(X)
```

```python
series_spread = pd.Series(spread, index=btc.index)

plt.figure(figsize=(16, 4))
plt.title('Spread between BTC and ETH')
plt.plot(series_spread, linewidth=1)
plt.show()
```
